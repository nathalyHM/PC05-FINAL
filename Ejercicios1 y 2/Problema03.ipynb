```python
import pandas as pd
df_airbnb = pd.read_csv("../data/airbnb.csv")
```

```python
df_airbnb.head(50)
```

```python
df_airbnb = df_airbnb[df_airbnb['price'] <= 50]
```

```python
Habitacion_compartida = df_airbnb[df_airbnb['room_type'] == 'Shared room']
```

```python
Habitacion_compartida_sorted = Habitacion_compartida.sort_values(by='overall_satisfaction', ascending=False)
```

```python
top_10_Habitacion_compartida = Habitacion_compartida_sorted.head(10)
```
